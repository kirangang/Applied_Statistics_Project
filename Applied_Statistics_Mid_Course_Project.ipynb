# Calculating Central Tendencies - Mean, Median and Mode

Since there are multiple variables it is better to have a selected list of numeric variables for calculating the central tendencies and dispersion.

```python
# 1. Mean

def calculate_mean(data, selected_columns):
    # Calculate mean
    mean_values = data[selected_columns].mean()

    result_data = pd.DataFrame({
        'Category': selected_columns,
        'Mean': mean_values.values,
        })

    return result_data

# Selected list
selected_list = ['Age', 'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts','MntSweetProducts', 'NumDealsPurchases', 'MntGoldProds',
                 'NumWebPurchases', 'NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']

result_data = calculate_mean(data, selected_list)

# Display the result
print(result_data.to_string(index=False))

```

```python
# 2. Median

def calculate_median(data, selected_columns):
    # Calculate median
    median_values = data[selected_columns].median()

    result_data = pd.DataFrame({
        'Category': selected_columns,
        'Median': median_values.values,
        })

    return result_data

# Selected list
selected_list = ['Age', 'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts','MntSweetProducts', 'NumDealsPurchases', 'MntGoldProds',
                 'NumWebPurchases', 'NumCatalogPurchases','NumStorePurchases','NumWebVisitsMonth']

result_data = calculate_median(data, selected_list)

# Display the result
print(result_data.to_string(index=False))
```

```python
# 3. Mode

def calculate_modes(data, selected_columns):
    modes = {}

    for variable in selected_columns:
        modes[variable] = data[variable].mode().values[0]

    return pd.Series(modes)

selected_list = ['Complain', 'Response', 'Income', 'Age', 'Kidhome', 'Teenhome', 'MntWines', 'NumWebVisitsMonth', 'Education', 'Marital_Status']

result_modes = calculate_modes(data, selected_list)
print(f'Mode of selected columns in the dataset are:\n{result_modes}')
```

# Measures of Dispersion - Variance and Standard Deviation

```python
# 1. Variance

def calculate_variance(data, selected_columns):
    # Calculate variance
    variance_values = data[selected_columns].var()

    result_data = pd.DataFrame({
        'Category': selected_columns,
        'Variance': variance_values.values,
        })

    return result_data

# Selected list
selected_list = ['Age', 'Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
                 'NumDealsPurchases', 'MntGoldProds', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

result_data = calculate_variance(data, selected_list)

# Display the result
print(result_data.to_string(index=False))
```

# Handling Outliers Using Boxplot

```python
# We will plot a box plot to detect outliers

def plot_boxplots(data, key_variables):
    # Set the number of rows and columns for subplots
    num_rows = (len(key_variables) + 2) // 3
    num_cols = 3

    # Create subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5*num_rows))
    fig.suptitle('Boxplots for Key Variables')

    # Flatten the axes array
    axes = axes.flatten()

    # Plot each numeric variable in a separate subplot
    for i, variable in enumerate(key_variables):
        sns.boxplot(x=data[variable], ax=axes[i])
        axes[i].set_title(variable)

    # Adjust layout
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])

    # Display the subplots
    plt.show()

key_variables = ['Income', 'Age', 'MntWines', 'MntFruits', 'MntMeatProducts', 'Recency']

plot_boxplots(data, key_variables)
```

###There are outliers present in various columns in our dataset, so we will remove those by IQR method.

```python
# Using IQR Method

# Function to remove outliers
def remove_outliers(data, columns):

    for column in columns:
        Q1 = data[column].quantile(0.20)
        Q3 = data[column].quantile(0.80)
        IQR = Q3 - Q1

        # Define the upper and lower bounds for outlier removal
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove outliers
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return data

columns_with_outliers = ['Income', 'Age', 'MntWines', 'MntFruits', 'MntMeatProducts']

# Remove outliers from specified columns
df_new = remove_outliers(data, columns_with_outliers)
```

####We will now check if outliers are cleaned or not.

## Probability and expected value calculation for normal distribution

```python
# 1. Let's calculate first for Income :

import numpy as np
from scipy.stats import norm

def calculate_probabilities_and_expectation(data, variable, threshold1, threshold2):
    # Calculate mean and standard deviation
    mean_val = np.mean(data[variable])
    std_dev_val = np.std(data[variable])

    # Calculate probabilities
    prob_less_than_threshold1 = norm.cdf(threshold1, loc=mean_val, scale=std_dev_val)
    prob_greater_than_threshold2 = 1 - norm.cdf(threshold2, loc=mean_val, scale=std_dev_val)

    # Calculate expected value (mean)
    expected_val = mean_val

    # Print results
    print(f"Variable: {variable}")
    print(f"Probability of {variable} less than {threshold1:.2f}: {prob_less_than_threshold1:.4f}")
    print(f"Probability of {variable} greater than {threshold2:.2f}: {prob_greater_than_threshold2:.4f}")
    print(f"Expected {variable}: {expected_val:.2f}\n")

calculate_probabilities_and_expectation(df_new, 'Income', 70000, 80000)

```

## Probability and expected value calculation for binomial distribution

```python
# Let's calculate probability and expected values for Response (if customer accepted the offer in the last campaign or not )
# and for Complain (complains made in last 2 years) variable

from scipy.stats import binom

def calculate_probabilities_and_expectation_binomial(data, variable):
    # Number of trials
    n_trials = len(data[variable])

    # Probability of success
    p_success = np.mean(data[variable])

    # Number of successes
    k_success = np.sum(data[variable] == 1)

    # Calculate probabilities using the PMF
    prob_success_1 = binom.pmf(k_success, n_trials, p_success)

    # Calculate expected value (mean)
    expected_value = n_trials * p_success

    # Print results
    print(f"Variable: {variable}")
    print(f"Probability of {variable} being 1: {prob_success_1:.4f}")
    print(f"Expected value for {variable}: {expected_value:.2f}\n")

calculate_probabilities_and_expectation_binomial(df_new, 'Response')
calculate_probabilities_and_expectation_binomial(df_new, 'Complain')
```

##2. Let's now segment customers based on total spending and income of customers.

```python
from sklearn.cluster import KMeans

def elbow_method_graph(df, variables, max_clusters=10):

    # Creating a new column TotalSpending for clustering
    df['TotalSpending'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

    # Applying k-means with different number of clusters
    inertia = []  # List to store the inertia for each k
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=0).fit(df[variables])
        inertia.append(kmeans.inertia_)

    # Plotting the Elbow Method Graph
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_clusters + 1), inertia, marker='o')
    plt.title("Elbow Method")
    plt.xlabel("Number of clusters")
    plt.ylabel("Inertia")
    plt.xticks(range(1, max_clusters + 1))
    plt.show()

key_variables_to_cluster = ['Income', 'TotalSpending']
elbow_method_graph(df_new, key_variables_to_cluster)
```

### From above plot we can see that optimal number of clusters should be 3

```python
# Function to plot clusters based on Income vs TotalSpending
def kmeans_clustering_and_visualization(df, features_for_clustering, num_clusters=3):

    # Fit KMeans model
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    df['Cluster1'] = kmeans.fit_predict(df[features_for_clustering])

    # Visualize the clusters based on 'TotalSpending' and 'Income'
    plt.figure(figsize=(12, 6))
    plt.scatter(df[df['Cluster1'] == 0]['Income'], df[df['Cluster1'] == 0]['TotalSpending'], s=80, c='red', label='Low Income, Low Spenders')
    plt.scatter(df[df['Cluster1'] == 1]['Income'], df[df['Cluster1'] == 1]['TotalSpending'], s=80, c='blue', label='Medium Income, Moderate Spenders')
    plt.scatter(df[df['Cluster1'] == 2]['Income'], df[df['Cluster1'] == 2]['TotalSpending'], s=80, c='green', label='High Income, High Spenders')

    plt.title('Clusters of Customers based on Total Spending vs Income')
    plt.xlabel('Annual Income')
    plt.ylabel('Total Spending')
    plt.legend()
    plt.show()

features_for_clustering = ['Income', 'TotalSpending']
kmeans_clustering_and_visualization(df_new, features_for_clustering)


```

Here we can observe that a majority of customers are of high income and high spenders, followed by low income and low spending habits.
